### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
s1=[int(i) for i in input("enter keys:").split()];
s=[int(i) for i in input("enter values:").split()];
k2=[int(i) for i in input("enter keys:").split()];
k=[int(i) for i in input("enter values:").split()];
d1=dict(zip(s1,s));
d2=dict(zip(k2,k));
ans=d2.copy();
for k,v in d1.items():
    if k in ans.keys():
        ans[k]+=v;
    else:
        ans[k]=v;
print(ans)
````
