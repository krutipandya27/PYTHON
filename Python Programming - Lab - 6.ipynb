### 17) WAP to create a list of common elements from given two lists.

```python
l1=[int(i) for i in input("enter value:").split()]
l2=[int(i) for i in input("enter value:").split()]
# l3=list(set(l2).intersection(set(l1)))
# print(l3)
l=[]
if len(l1)>len(l2):
    high=len(l2)
else:
    high=len(l1)
for i in range(high):
    if l1[i] in l2:
        l.append(l1[i])
        l2.remove(l1[i])
print(l)
```
