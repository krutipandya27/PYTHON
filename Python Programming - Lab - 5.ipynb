### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s=input("enter string")
count=1;
ans=""
for i in range(0,len(s)-1):
    if s[i]==s[i+1]:
        count+=1;
    else:
        ans+=s[i]+str(count)
        count=1
ans+=s[i+1]+str(count)
print(ans)  
```
