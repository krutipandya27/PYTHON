### 14) WAP to check whether the given number is Palindrome or not.

```python
m=int(input("enter num:"))
rem=0;
ans=0;
n=m;
while m!=0:
    rem=m%10;
    ans=ans*10+rem;
    m=m//10;

print(ans);

if ans==n:
    print("palindrome")
else:
    print("not palindrome")
```
